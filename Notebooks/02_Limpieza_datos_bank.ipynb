```python
def valores_minus (df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].str.lower()
    return df
```

```python
df_limpios = valores_minus(df_limpios)
```

```python
df_limpios.sample(3)
```

```python
def convertir_columnas (df):
    for col in df.columns:
        for dtype in [float, int]:
            try:
                df[col] = df[col].astype(dtype)
            except:
                pass
    return df
```

```python
df_limpios = convertir_columnas(df_limpios)
```

```python
columnas_comas = ['cons.price.idx', 'cons.conf.idx',
       'euribor3m', 'nr.employed']
```

```python
def puntos_por_comas (df, lista_col):
    for col in lista_col:
        df[col] = df[col].str.replace(",",".")
    return df
```

```python
df_limpios = puntos_por_comas(df_limpios, columnas_comas)
```

```python
df_limpios.sample(3)
```

```python
meses_es = {"enero": "01", "febrero":"02", "marzo":"03", "abril": "04", "mayo":"05", "junio":"06", "julio": "07", "agosto":"08", "septiembre":"09", "octubre":"10", "noviembre":"11","diciembre":"12"}

def traducir_fecha(fecha_str):
    if pd.isna(fecha_str):
        return fecha_str
    fecha_str = fecha_str.lower()
    for mes, num in meses_es.items():
        fecha_str = fecha_str.replace(mes,num)
    return fecha_str
```

```python
df_limpios['date'] = df_limpios['date'].apply(traducir_fecha)
```

```python
df_limpios['date'] = pd.to_datetime(df_limpios['date'], format="%d-%m-%Y", errors="coerce")
```
